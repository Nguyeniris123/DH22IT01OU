#Bài tập 1: Tải và Chuyển đổi Dữ liệu Iris


##Yêu cầu 1.1: Tải dữ liệu



```python
# Them cac thu vien can thiet
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris_data.feature_names
```

##Yêu cầu 1.2: Tạo DataFrame ban đầu



```python
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df
```

##Yêu cầu 1.3: Thêm thông tin về loài hoa (Target)


```python
df['species'] = iris_data.target
df
```
