#Bài 1: Tiền xử lý và Khám phá Dữ liệu (Data Preprocessing & EDA)


##1. Hợp nhất dữ liệu:


```python
import pandas as pd
url_diem_thi2022 = "/content/drive/MyDrive/Colab Notebooks/Data/diem_thi_thpt_2022.csv"
url_diem_thi2023 = "/content/drive/MyDrive/Colab Notebooks/Data/diem_thi_thpt_2023.csv"
url_diem_thi2024 = "/content/drive/MyDrive/Colab Notebooks/Data/diem_thi_thpt_2024.csv"

df_diem_thi2022 = pd.read_csv(url_diem_thi2022)
df_diem_thi2022.head(5)
```

Điểm thi THPT 2023

```python
df_diem_thi2023 = pd.read_csv(url_diem_thi2023)
df_diem_thi2023.head(5)
```

Điểm thi THPT 2024

```python
df_diem_thi2024 = pd.read_csv(url_diem_thi2024)
df_diem_thi2024.head(5)
```

```python
# Thêm cột NamThi
df_diem_thi2022['NamThi'] = 2022
df_diem_thi2023['NamThi'] = 2023
df_diem_thi2024['NamThi'] = 2024
```

```python
# Xóa dữ liệu trùng lặp
df_diem_thi2022.drop_duplicates(inplace=True)
df_diem_thi2023.drop_duplicates(inplace=True)
df_diem_thi2024.drop_duplicates(inplace=True)
```

```python
# Xóa trùng dựa trên số báo danh (Giữ lại dòng đầu tiên)
df_diem_thi2022.drop_duplicates(subset='sbd', keep='first', inplace=True)
df_diem_thi2023.drop_duplicates(subset='sbd', keep='first', inplace=True)
df_diem_thi2024.drop_duplicates(subset='sbd', keep='first', inplace=True)

# Xóa cột mã ngoại ngữ của df_diem_thi2023, df_diem_thi2024
df_diem_thi2023.drop('ma_ngoai_ngu', axis=1, inplace=True, errors='ignore')
df_diem_thi2024.drop('ma_ngoai_ngu', axis=1, inplace=True, errors='ignore')
```

```python
# Concatenate the three dataframes
df_diem_thi_combined = pd.concat([df_diem_thi2022, df_diem_thi2023, df_diem_thi2024], ignore_index=True)
df_diem_thi_combined
```

```python
df_diem_thi_combined.info()
```

##2. Xử lý dữ liệu khuyết thiếu (Missing Values):


```python
df_diem_thi_combined.rename(columns={
    'sbd': 'ID',
    'toan': 'Toan',
    'ngu_van': 'Ngu Van',
    'ngoai_ngu': 'Ngoai Ngu',
    'dia_li': 'Dia Li',
    'lich_su': 'Lich Su',
    'gdcd': 'GDCD',
    'sinh_hoc': 'Sinh Hoc',
    'hoa_hoc': 'Hoa Hoc',
    'vat_li': 'Vat Li',
    'ma_ngoai_ngu': 'Ma Ngoai Ngu'
}, inplace=True)
df_diem_thi_combined.info()
```

```python
#1.3 Liet ke cot tinh diem
score_cols = ['Toan', 'Ngu Van', 'Ngoai Ngu', 'Dia Li', 'Lich Su', 'GDCD', 'Sinh Hoc', 'Hoa Hoc', 'Vat Li']
```

```python
# Dien gia tri thieu NaN bang 0 trong df_diem_thi_combined
df_diem_thi_combined[score_cols] = df_diem_thi_combined[score_cols].fillna(0)
```

```python
df_diem_thi_combined
```

##3. Tạo thuộc tính mới (Feature Engineering):

```python
df_diem_thi_combined['ID'] = df_diem_thi_combined['ID'].astype(str).str.zfill(8)
df_diem_thi_combined['MaTinh'] = df_diem_thi_combined['ID'].str[:2]

ma_tinh_dict = {
    '01': 'Hà Nội', '02': 'TP. Hồ Chí Minh', '03': 'Hải Phòng', '04': 'Đà Nẵng', '05': 'Hà Giang',
    '06': 'Cao Bằng', '07': 'Lai Châu', '08': 'Lào Cai', '09': 'Tuyên Quang', '10': 'Lạng Sơn',
    '11': 'Bắc Kạn', '12': 'Thái Nguyên', '13': 'Yên Bái', '14': 'Sơn La', '15': 'Phú Thọ',
    '16': 'Vĩnh Phúc', '17': 'Quảng Ninh', '18': 'Bắc Giang', '19': 'Bắc Ninh', '21': 'Hải Dương',
    '22': 'Hưng Yên', '23': 'Hòa Bình', '24': 'Hà Nam', '25': 'Nam Định', '26': 'Thái Bình',
    '27': 'Ninh Bình', '28': 'Thanh Hóa', '29': 'Nghệ An', '30': 'Hà Tĩnh', '31': 'Quảng Bình',
    '32': 'Quảng Trị', '33': 'Thừa Thiên Huế', '34': 'Quảng Nam', '35': 'Quảng Ngãi', '36': 'Kon Tum',
    '37': 'Bình Định', '38': 'Gia Lai', '39': 'Phú Yên', '40': 'Đăk Lăk', '41': 'Khánh Hòa',
    '42': 'Lâm Đồng', '43': 'Bình Phước', '44': 'Bình Dương', '45': 'Ninh Thuận', '46': 'Tây Ninh',
    '47': 'Bình Thuận', '48': 'Đồng Nai', '49': 'Long An', '50': 'Đồng Tháp', '51': 'An Giang',
    '52': 'Bà Rịa - Vũng Tàu', '53': 'Tiền Giang', '54': 'Kiên Giang', '55': 'Cần Thơ',
    '56': 'Bến Tre', '57': 'Vĩnh Long', '58': 'Trà Vinh', '59': 'Sóc Trăng', '60': 'Bạc Liêu',
    '61': 'Cà Mau', '62': 'Điện Biên', '63': 'Đăk Nông', '64': 'Hậu Giang',
}

df_diem_thi_combined['TenTinh'] = df_diem_thi_combined['MaTinh'].map(ma_tinh_dict)
df_diem_thi_combined
```

```python
df_diem_thi_combined['KhoiA'] = df_diem_thi_combined['Toan'] + df_diem_thi_combined['Vat Li'] + df_diem_thi_combined['Hoa Hoc']
df_diem_thi_combined['KhoiA1'] = df_diem_thi_combined['Toan'] + df_diem_thi_combined['Vat Li'] + df_diem_thi_combined['Ngoai Ngu']
df_diem_thi_combined['KhoiB'] = df_diem_thi_combined['Toan'] + df_diem_thi_combined['Sinh Hoc'] + df_diem_thi_combined['Hoa Hoc']
df_diem_thi_combined['KhoiC'] = df_diem_thi_combined['Ngu Van'] + df_diem_thi_combined['Dia Li'] + df_diem_thi_combined['Lich Su']
df_diem_thi_combined['KhoiD'] = df_diem_thi_combined['Ngoai Ngu'] + df_diem_thi_combined['Toan'] + df_diem_thi_combined['Ngu Van']
df_diem_thi_combined.head()
```

```python
df_diem_thi_combined.to_csv('/content/drive/MyDrive/Colab Notebooks/Data/df_diem_thi_combined.csv', index=False)
```

##2. Phân tích theo Tỉnh/Thành (VD: năm 2024):


```python
df_diem_thi2024 = df_diem_thi_combined[df_diem_thi_combined['NamThi'] == 2024].copy()
avg_cols = score_cols + ['KhoiA', 'KhoiA1', 'KhoiB', 'KhoiC', 'KhoiD']
```

```python
province_avg_scores = df_diem_thi2024.groupby(['MaTinh', 'TenTinh'])[avg_cols].mean().round(2)
province_avg_scores
```

```python
province_ranking = province_avg_scores.reset_index()
print("\nTop 5 tỉnh có điểm trung bình khối A cao nhất: ")
top5_KhoiA = province_ranking[["TenTinh", "KhoiA"]].sort_values(by="KhoiA", ascending=False).head(5).set_index("TenTinh")
print(top5_KhoiA)

print("\nTop 5 tỉnh có điểm trung bình khối B cao nhất: ")
top5_KhoiB = province_ranking[["TenTinh", "KhoiB"]].sort_values(by="KhoiB", ascending=False).head(5).set_index("TenTinh")
print(top5_KhoiB)

print("\nTop 5 tỉnh có điểm trung bình khối C cao nhất: ")
top5_KhoiC = province_ranking[["TenTinh", "KhoiC"]].sort_values(by="KhoiC", ascending=False).head(5).set_index("TenTinh")
print(top5_KhoiC)

print("\nTop 5 tỉnh có điểm trung bình khối D cao nhất: ")
top5_KhoiD = province_ranking[["TenTinh", "KhoiD"]].sort_values(by="KhoiD", ascending=False).head(5).set_index("TenTinh")
print(top5_KhoiD)
```

## 3. Trực quan hóa Địa lý (Geospatial Visualization):


```python
import geopandas as gpd
vn_map = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Data/diaphantinh.geojson")
vn_map.head()
```

```python
# vn_map[['gid', 'ten_tinh']].to_csv('test.csv')
```

```python
correction = {
    'Bà Rịa -Vũng Tàu': 'Bà Rịa - Vũng Tàu',
    'Cần Thơn': 'Cần Thơ',
    'Kien Giang': 'Cà Mau',
    'Đăk Lăk': 'Kiên Giang',
    'Quản Bình': 'Quảng Bình'
}

vn_map['ten_tinh'] = vn_map['ten_tinh'].replace(correction)
```

```python
vn_map.head()
```

```python
# merge dl dựa trên tên tỉnh
vn_map_scores = vn_map.merge(
    province_avg_scores,
    left_on='ten_tinh',
    right_on='TenTinh',
    how='left'
)

num_total_provinces = len(vn_map_scores)
num_match_provinces = vn_map_scores['KhoiA'].notna().sum()
print(f"{num_total_provinces} khớp {num_match_provinces}")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 16))


vn_map_scores.plot(
   column='KhoiA',          # Dữ liệu để tô màu (ví dụ: điểm khối A)
   cmap='Blues',            # Chọn một bảng màu khác, ví dụ: 'Blues'
   linewidth=0.7,
   ax=ax,
   edgecolor='0.6',
   legend=True,
   missing_kwds={          # Định dạng cho các tỉnh không có dữ liệu
       "color": "lightgrey",
       "edgecolor": "red",
       "hatch": "///",
       "label": "Missing values",
   },
   legend_kwds={
       'label': "Điểm trung bình khối A năm 2024",
       'orientation': "horizontal"
   }
)


# Thêm tên tỉnh vào bản đồ
for idx, row in vn_map_scores.iterrows():
   if row['KhoiA'] > 12 and not row['geometry'].is_empty: # vn_map_scores['KhoiA'].median()
       x, y = row['geometry'].centroid.coords[0]
       ax.text(x, y, row['ten_tinh'], fontsize=6, ha='center', va='center')


ax.axis('off')
ax.set_title('Bản đồ Phân bố Điểm trung bình Khối A các tỉnh (2024)',
             fontdict={'fontsize': '18', 'fontweight': 'bold'})


plt.show()
```
