#Phần 1: Dữ liệu Iris


##Yêu cầu 1.1: Tải, Khám phá và Chuẩn bị Dữ liệu


```python
from sklearn.datasets import load_iris

#Tai va tao dataframe
iris = load_iris()
iris
```

```python
import pandas as pd
x_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
y_iris_true = iris.target
x_iris.head(5)
```

```python
y_iris_true
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(x_iris)
plt.suptitle("Phan bo cac dac trung trong du lieu Iris", y=1.02)
plt.show()
```

```python
# Chuan hoa bang StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_iris_scaled = scaler.fit_transform(x_iris)

print(x_iris_scaled[:5])
```

Chuẩn hóa dữ liệu đảm bảo rằng tất cả các đặc trưng đều đóng góp vào việc tính toán khoảng cách một cách công bằng, ngăn chặn việc các đặc trưng có thang đo lớn hơn áp đảo các đặc trưng khác. Điều này giúp các thuật toán gom cụm dựa trên khoảng cách tìm thấy các cụm thực sự dựa trên cấu trúc tổng thể của dữ liệu thay vì chỉ dựa trên sự khác biệt về thang đo.

Trong trường hợp dữ liệu Iris, các đặc trưng đều có đơn vị là cm, nhưng phạm vi giá trị của chúng khác nhau (ví dụ: sepal length từ 4.3 đến 7.9, petal width từ 0.1 đến 2.5). Mặc dù sự khác biệt về thang đo không quá lớn, việc chuẩn hóa vẫn giúp đảm bảo rằng cả bốn đặc trưng đều có trọng số ngang nhau trong quá trình gom cụm.

##Yêu cầu 1.2: Lựa chọn số cụm k tối ưu cho K-Means


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Phuong phap Elbow
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(x_iris_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.title('Phương pháp Elbow')
plt.xlabel('Số cụm (k)')
plt.ylabel('SSE')
plt.grid(True)
plt.show()
```

```python
silhouette_coefficients = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(x_iris_scaled)
    score = silhouette_score(x_iris_scaled, kmeans.labels_)
    silhouette_coefficients.append(score)

plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_coefficients, marker='o')
plt.title('Hệ số Sillhouette cho các giá trị k')
plt.xlabel('Số cụm (k)')
plt.ylabel('Hệ số Sillhouette trung bình')
plt.grid(True)
plt.show()
```

Phương pháp Elbow: Biểu đồ cho thấy điểm "khuỷu tay" rõ rệt nhất là tại k = 3. Tại điểm này, sự giảm SSE (Sum of Squared Errors) bắt đầu chậm lại, cho thấy việc tăng thêm số cụm không mang lại nhiều cải thiện đáng kể về mặt giảm tổng bình phương khoảng cách từ các điểm đến tâm cụm.

Hệ số Sillhouette: Biểu đồ cho thấy hệ số Silhouette cao nhất đạt được tại k = 3. Hệ số Silhouette đo lường mức độ tương đồng của một đối tượng với cụm của chính nó so với các cụm khác. Giá trị cao nhất cho thấy các cụm được phân tách rõ ràng và các điểm nằm chặt chẽ trong cụm của chúng.

Cả hai phương pháp đều chỉ ra rằng k = 3 là giá trị tối ưu cho số cụm trên tập dữ liệu Iris này. Điều này cũng phù hợp với thông tin ban đầu về tập dữ liệu Iris có 3 loài hoa khác nhau.

##Yêu cầu 1.3: So sánh các Thuật toán Gom cụm


```python
from sklearn.cluster import KMeans
# Gom cum Kmeans voi so cum 3
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(x_iris_scaled)
```

```python
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
# Gom cum phan cap
linked = linkage(x_iris_scaled, method="ward")

plt.figure(figsize=(12, 8))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendoeogram cho dl Iris (Ward\'s Method) ')
plt.xlabel('Chi so diem dl')
plt.ylabel('Khoang cach')
plt.show()

# Cat Dendrogram de chi lay 3 cum
agg_cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_agg = agg_cluster.fit_predict(x_iris_scaled)
```

```python
dbscan = DBSCAN(eps=0.9, min_samples=5)
y_dbscan = dbscan.fit_predict(x_iris_scaled)
```

```python
# Truc quan hoa va so sanh ket qua
fig, axes = plt.subplots(1,4, figsize=(20,5), sharey=True)

#lay 2 dac trung quan trong nhat de ve petal lenght, va petal width

x_axis = 2
y_axis = 3

# Bieu do nhan that
axes[0].scatter(x_iris_scaled[:, x_axis], x_iris_scaled[:, y_axis], c=y_iris_true, cmap='viridis')
axes[0].set_title("Ground Truth (Nhan that)")
axes[0].set_xlabel(iris.feature_names[x_axis])
axes[0].set_ylabel(iris.feature_names[y_axis])

# Bieu do KMeans
axes[1].scatter(x_iris_scaled[:, x_axis], x_iris_scaled[:, y_axis], c=y_kmeans, cmap='viridis')
axes[1].set_title("Kmeans Clustering (k=3)")
axes[1].set_xlabel(iris.feature_names[x_axis])
axes[1].set_ylabel(iris.feature_names[y_axis])

# Bieu do Agglomerative
axes[2].scatter(x_iris_scaled[:, x_axis], x_iris_scaled[:, y_axis], c=y_agg, cmap='viridis')
axes[2].set_title("Hierarchy Clustering")
axes[2].set_xlabel(iris.feature_names[x_axis])
axes[2].set_ylabel(iris.feature_names[y_axis])

# Bieu do DBSCAN
axes[3].scatter(x_iris_scaled[:, x_axis], x_iris_scaled[:, y_axis], c=y_dbscan, cmap='viridis')
axes[3].set_title("DBSCAN Clustering")
axes[3].set_xlabel(iris.feature_names[x_axis])
axes[3].set_ylabel(iris.feature_names[y_axis])

plt.suptitle('So sanh ket qua cua cac thuat toan gom cum')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Không có thuật toán nào hiệu quả hoàn toàn. Tùy từng trường hợp Ta cần thử từng thuật toán để đánh giá

KMeans (k=3): Biểu đồ cho thấy KMeans với k=3 đã phân chia dữ liệu thành ba cụm khá rõ ràng, tương đối giống với phân bố của nhãn thật. Tuy nhiên, vẫn có một số điểm bị gán nhầm cụm, đặc biệt là giữa hai cụm phía trên.
Agglomerative Clustering (Hierarchical Clustering) với phương pháp Ward và 3 cụm: Kết quả của gom cụm phân cấp cũng cho thấy ba cụm tương tự KMeans. Mức độ phân tách và sự tương đồng với nhãn thật cũng khá tốt, nhưng vẫn có sự chồng chéo và gán nhầm cụm ở ranh giới giữa các cụm.
DBSCAN (với eps=0.9, min_samples=5): Biểu đồ của DBSCAN cho thấy thuật toán này đã tìm được một cụm chính và một số điểm được coi là nhiễu (màu tím). DBSCAN dường như gặp khó khăn trong việc phân tách rõ ràng ba nhóm dựa trên các tham số đã chọn, có thể do sự chồng chéo giữa các lớp trong dữ liệu Iris hoặc cần tinh chỉnh tham số eps và min_samples.

Cả KMeans và Agglomerative Clustering với số cụm k=3 đều cho kết quả gom cụm khá tốt và tương đồng với nhãn thật trên bộ dữ liệu Iris, phản ánh cấu trúc 3 lớp của dữ liệu.
DBSCAN với các tham số hiện tại không hiệu quả bằng KMeans và Agglomerative Clustering trong việc phân tách ba lớp của dữ liệu Iris, nó có xu hướng gom một số lớp lại thành một cụm lớn và coi các điểm còn lại là nhiễu.

# Phần 2: Phân tích Dữ liệu điểm thi THPT Quốc gia 2024


##Yêu cầu 2.1: Chuẩn bị Dữ liệu Điểm thi (2.0 điểm)


```python
import pandas as pd
url_diem_thi = "/content/drive/MyDrive/Colab Notebooks/Data/df_diem_thi_combined.csv"

df_2024 = pd.read_csv(url_diem_thi)
df_2024.info()
```

```python
feature_cols = ['Toan', 'Ngu Van', 'Ngoai Ngu', 'Vat Li', 'Hoa Hoc', 'Sinh Hoc', 'Lich Su', 'Dia Li', 'GDCD', 'KhoiA', 'KhoiD']
province_features = df_2024.groupby("TenTinh")[feature_cols].mean().round(2)

print("Dac trung diem thi cua cac tinh (5 tinh dau tien)")
province_features.head()
```

```python
#Chuan hoa dl
from sklearn.preprocessing import StandardScaler
scaler_province = StandardScaler()
province_features_scaled = scaler_province.fit_transform(province_features)
```

```python
from sklearn.metrics import silhouette_score
# Tim so cum k toi uu voi silhouette
silhouette_coefficients_province = []

k_range= range(2,11)
for k in k_range:
  kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
  kmeans.fit(province_features_scaled)
  score = silhouette_score(province_features_scaled, kmeans.labels_)
  silhouette_coefficients_province.append(score)\

plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_coefficients_province, marker='o')
plt.title('Hệ số Sillhouette cho các giá trị k')
plt.xlabel('Số cụm (k)')
plt.ylabel('Hệ số Sillhouette trung bình')
plt.grid(True)
plt.show()
```

```python
k_optimal = 4
#Chay Kmeans voi k toi uu
kmeans_province = KMeans(n_clusters=k_optimal, n_init=10, random_state=42)
province_features['Cluster'] = kmeans_province.fit_predict(province_features_scaled)
province_features.head()
```

```python
# Phan tich va dien giai cac cum
print(f"\n------- Phan tich dac trung Trung binh cua {k_optimal} cum tinh/thanh ------")
cluster_analysis = province_features.groupby('Cluster')[feature_cols].mean()
cluster_analysis
```

```python
# in ra cac tinh trong cum

for i in range(k_optimal):
  provinces_in_cluster = province_features[province_features['Cluster'] ==i].index.to_list()
  print(f"\nCum {i}: ")
  print(", ".join(provinces_in_cluster))
```
